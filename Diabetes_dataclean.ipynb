```python
import pandas as pd
import seaborn as sns
```

#### Reading CSV file

```python
df = pd.read_csv(r'diabetes_unclean.csv')
df
```

```python
df.head()
```

#### Checking number of Columns

```python
df.columns
```

#### Renaming incorrect Column name 'No_Pation'

```python
df1 = df.rename(columns = {"No_Pation":"Patient_no"})
df1
```

#### Checking for missing values

```python
df1.isnull().sum()
```

#### Replacing missing values in 'Hba1c' Column

```python
mean_value = df1['HbA1c'].mean()
mean_value
```

```python
df1['HbA1c'].fillna(mean_value, inplace = True)
df1
```

```python
df1.isnull().sum()
```

#### Dropping missing values in remaining columns

```python
df2 = df1.dropna()
df2.isnull().sum()
```

#### Checking for duplicate values

```python
df2.info()
```

```python
df2.groupby('CLASS')['CLASS'].agg("count")
```

#### Checking unique values in CLASS column

```python
df2['CLASS'].unique()
```

#### Cleaning the column, replacing unique values

```python
df2['CLASS'] = df2['CLASS'].str.replace("Y ","Y")
df2['CLASS'] = df2['CLASS'].str.replace("N ","N")
```

```python
df2['CLASS'].unique()
```

#### Checking for Outliers

```python
sns.boxplot(df2["Cr"])
```

#### Choose the maximum quantile to replace outlier

```python
max_cr = df2["Cr"].quantile(0.995)
max_cr
```

```python
df2[df2["Cr"] > max_cr]
```

####  Assigning new data to new dataset

```python
df3 = df2[df2["Cr"] < max_cr]
df3
```

```python
sns.boxplot(df3["Cr"])
```

#### Checking for Outliers for HbA1c column

```python
sns.boxplot(df3["HbA1c"])
```

#### Checking for Duplicate values

```python
df3.duplicated()
```

```python
df4 = df3.drop_duplicates()
```

```python
df4.duplicated()
```
